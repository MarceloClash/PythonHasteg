```python
import statistics as st

salarios =  [1555, 2000, 3000, 4222, 5222, 8222]
funcionarios = ['marcos', 'lucas', 'cleber','ana','joao', 'marcelo']

index = 0
for salario in salarios:
  print(f'O salário de {funcionarios[index]} é de R$ {salario}')
  index +=1
```

```python
media_salarial = st.mean(salarios)
salario_maior = max(salarios)
salario_menor = min(salarios)
salario_mediana = st.median(salarios)
salario_moda = st.mode(salarios)
salario_desvio_padrao = st.stdev(salarios)
salario_variancia = st.variance(salarios)
```

```python
print(f'O maior salario é de R$ {salario_maior}')
print(f'O menor salario é de R$ {salario_menor}')
print(f'A média salarial é de R$ {media_salarial:.2f}')
print(f'A mediana salarial é de R$ {salario_mediana}')
print(f'A moda salarial é de R$ {salario_moda:.2f}')


```

```python
# prompt: com os dados quero gerar um grafico

import statistics as st
import matplotlib.pyplot as plt

salarios =  [1555, 2000, 3000, 4222, 5222, 8222]
funcionarios = ['marcos', 'lucas', 'cleber','ana','joao', 'marcelo']

index = 0
for salario in salarios:
  print(f'O salário de {funcionarios[index]} é de R$ {salario}')
  index +=1

media_salarial = st.mean(salarios)
salario_maior = max(salarios)
salario_menor = min(salarios)
salario_mediana = st.median(salarios)
salario_moda = st.mode(salarios)
salario_desvio_padrao = st.stdev(salarios)
salario_variancia = st.variance(salarios)

print(f'O maior salario é de R$ {salario_maior}')
print(f'O menor salario é de R$ {salario_menor}')
print(f'A média salarial é de R$ {media_salarial:.2f}')
print(f'A mediana salarial é de R$ {salario_mediana}')
print(f'A moda salarial é de R$ {salario_moda:.2f}')

```

```python
#Plotando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(funcionarios, salarios, color='skyblue')
plt.xlabel("Funcionários")
plt.ylabel("Salários (R$)")
plt.title("Salários dos Funcionários")
plt.xticks(rotation=45, ha='right') # Rotaciona os nomes dos funcionários para melhor visualização
plt.tight_layout() # Ajusta o layout para evitar sobreposição de elementos
plt.show()
```
