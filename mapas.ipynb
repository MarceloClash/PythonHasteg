```python
!pip install folium # Instalando a biblioteca folium, caso não esteja instalada

import folium as fl # Importando a biblioteca folium e atribuindo a ela o apelido "fl"
from IPython.display import display # Importando a função "display" para exibir o mapa no notebook


# Definindo as coordenadas do centro do mapa (Berlim)
latitude = 52.5200
longitude = 13.4050
map_center = [latitude, longitude]

# Criando o objeto do mapa com o centro em Berlim e zoom inicial 15
mymap = fl.Map(location=map_center, zoom_start=15)

# Criando um marcador para Berlim
marker = fl.Marker(
    location=[latitude, longitude], # Coordenadas do marcador
    popup='Berlim', # Texto que aparece ao clicar no marcador
    icon=fl.Icon(color='red', icon='info-sign') # Ícone verde de informação
)

# Adicionando o marcador ao mapa
marker.add_to(mymap)

# Exibindo o mapa no notebook
display(mymap)
```

```python
import folium

# Crie um mapa centrado em São Paulo
mapa = folium.Map(location=[-23.55, -46.63], zoom_start=10)

# Adicione um círculo com popup HTML
folium.Circle(
    location=[-23.55, -46.63],
    radius=5000,
    popup="""
        <h1>São Paulo</h1>
        <p>A maior cidade do Brasil.</p>
        <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9a/Bandeira_de_S%C3%A3o_Paulo.svg/1200px-Bandeira_de_S%C3%A3o_Paulo.svg.png" width="200">
    """,
    fill=True,
    color="blue",
    fill_color="blue"
).add_to(mapa)

# Exiba o mapa
mapa
```
